```python

sk = "sakila"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+ sk
engine = create_engine(connection_string)
engine
```

```python
from sqlalchemy import text
with engine.connect() as connection:
    query = text('SELECT * FROM store')
    result = connection.execute(query)
    for row in result.mappings():
        print(row)
```

```python
with engine.connect() as connection:
    query = text(f'''SELECT * 
                     FROM customer
                     WHERE first_name LIKE "K%" AND active = 1''')
    result = connection.execute(query)
    df = pd.DataFrame(result.all())

df  
```

```python
with engine.connect() as connection:
    query = text(f'''SELECT COUNT(amount) 
                    FROM payment
                    WHERE amount > 10''')
    result = connection.execute(query)
    print(pd.DataFrame(result.all()))
```

```python
def payments_over_x(x):
    with engine.connect() as connection:
        txt = f'''SELECT COUNT(amount) AS number_of_payments
                        FROM payment
                        WHERE amount > {x} ;'''
        query = text(txt)
        result = connection.execute(query)
    return print(pd.DataFrame(result.all()))
```

```python
payments_over_x(5)
```

```python
def get_films_from_actor(first_name,last_name):
    with engine.connect() as connection:
        txt = f'''SELECT title
                FROM film
                INNER JOIN film_actor
                USING(film_id)
                INNER JOIN actor
                USING(actor_id) 
                WHERE first_name = {first_name} and last_name = {last_name};'''
        query = text(txt)
        result = connection.execute(query)
    return print(pd.DataFrame(result.all()))
```

```python
first_name = input()
last_name = input()
get_films_from_actor(first_name,last_name)
```
